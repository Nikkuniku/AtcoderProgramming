```python
import networkx as nx
import matplotlib.pyplot as plt 
G = nx.Graph()
# 頂点の追加
G.add_node(1)                
G.add_nodes_from([3, 4, 5])

# 辺の追加 (頂点も必要に応じて追加されます)
G.add_edge(1, 2)                                    
G.add_edges_from([(1, 3), (2, 5), (3, 4), (4, 5),(1,5)])

nx.draw_networkx(G)
plt.show()
```
